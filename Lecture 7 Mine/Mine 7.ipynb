```python
import numpy as np
import matplotlib.pyplot as plt

Length_meters = 100
H = 1000
span_weight_newtons = 50
num_points = 20

x = np.linspace(0, Length_meters, num_points)

def displacement(x):
    return -(H / span_weight_newtons) * (
        np.cosh((-span_weight_newtons / H) * (x - (Length_meters / 2))) -
        np.cosh((span_weight_newtons * Length_meters) / (2 * H))
    )

y = -displacement(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Cable Sag')
plt.scatter(x, y, color='red', marker='x', label='Vertices')
plt.xlabel('Horizontal Distance (m)')
plt.ylabel('Vertical Position (m)')
plt.title('Cable Sag Profile')
plt.legend()
plt.grid()
plt.show()

max_displacement = displacement(Length_meters / 2)
("Max displacement", max_displacement)
```

```python
import math

L_cat=0
for i in np.arange(0,len(x)-1):
    dx = abs(x[i+1]-x[i])
    dy = abs(y[i+1]-y[i])
    L_cat += math.sqrt(dx**2 + dy**2)   

# Cable length
L_cat
```

```python
# Vetical pulling force must support half the weight of the cable
V = 0.5*(L_cat*span_weight_newtons)

# The tension is the sum of the H and V vectors. This is just it's magnitude.
T_max = math.sqrt(H**2 + V**2)

print(f'The vertical reaction is {round(V,2)} N')
print(f'The maximum cable tension at supports is {round(T_max,2)} N')
```
